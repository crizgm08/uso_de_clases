Usar el return e imput con clases
siempre que use imput y quiera ingresar un numero con este, debo poner antes int, ya que el imput siempre leera todo en formato string

```python
class calculadoraurgencias:
    
    def __init__(self, region, consultas_ayer, consultas_hoy):
        self.region= region
        self.consultasyesterdays= consultas_ayer
        self.consultastoday= consultas_hoy

    def variaciondeurgencias(self):
        diferencia= self.consultasyesterdays-self.consultastoday
        return(diferencia)


#USO

mi_calculadora= calculadoraurgencias("Santiago",12,7) ### con estos valores la clase trabaja en su totalidad y lo que haga dentro de ella lo guardara la variable mi_calculadora

resultado=mi_calculadora.variaciondeurgencias()  ## y para extraer un valor particular debo usar la variable en donde se guardo todo y seguido poner un "." y la funcion especifica quie quiero extraer sus resultados

print(f"La variacion de urgencias entre today and yesterdays es: {resultado}")
```

```python
class registrourgencia:
    def __init__(self, region, registro):
        self.region= region
        self.registro= registro


## difencias entre input

mi_region= input("Ingrese su regresion: ")
mi_registro= int(input("Ingrese su numero de registro: "))

resultado1=registrourgencia(mi_region,mi_registro)

print(f"Su region es:{resultado1.region}| y su numero de registro : {resultado1.registro}")
```
